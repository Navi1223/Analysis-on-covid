```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Simplify the column names

```python
df.columns
```

```python
df.columns=['State','Total_cases','Active','Deaths','Active_Ratio','Death_Ratio']
df.head()
```

## In which state active ratio are maximum

```python
df.groupby(['State'])['Active_Ratio'].max()
```

```python
df.groupby(['State'])['Active_Ratio'].sum().sort_values(ascending=False)
```

```python
df.sort_values(by='Active_Ratio',ascending=False)
```

## Plot active ratio

```python
plt.figure(figsize=(10,4),dpi=200)
plt.plot(df['State'],df['Active_Ratio'],color='red',marker='o',markersize=10)
plt.xticks(rotation=90,fontsize=11)
plt.yticks(fontsize=13.5)
plt.xlabel('State',fontsize=15)
plt.ylabel('Active Ratio',fontsize=15)
plt.title('Line plot for Active Ratio',fontsize=15)
plt.show()
```

## Find the total cases in state

```python
q=df.sort_values(by='Total_cases',ascending=False)
q=q.head(10)
q
```

```python
plt.pie(q['Total_cases'],labels=q['State'],autopct='%0.0f%%');
```

```python
plt.bar(q['Total_cases'],q['State'],color='red');
plt.show()
```

## Plot the death ratio

```python
plt.figure(figsize=(10,4),dpi=200)
plt.plot(df['State'],df['Death_Ratio'],color='red',marker='o',markersize=10)
plt.xticks(rotation=90,fontsize=11)
plt.yticks(fontsize=13.5)
plt.xlabel('State',fontsize=15)
plt.ylabel('Death Ratio',fontsize=15)
plt.title('Line plot for Death Ratio',fontsize=15)
plt.show()
```

## Active and Death cases


```python
plt.figure(figsize=(9,6),dpi=200)

plt.title('Line plot for Active Ratio and Death Ratio',fontsize=15)

# active ratio
plt.plot(df['State'],df['Active_Ratio'],color='blue',marker='o',label='Active cases')
plt.xticks(rotation=90,fontsize=12.5)
plt.yticks(fontsize=12.5)
plt.xlabel('State',fontsize=15)
plt.ylabel('Active Ratio',fontsize=15)

# death ratio
plt.plot(df['State'],df['Death_Ratio'],color='k',marker='*',label='Death cases')

plt.legend()
plt.show()
```

## Add column Recovered_cases

```python
df['recovered_cases']=df['Total_cases']-(df['Active']+df['Deaths'])
df.head()
```

```python
df['recovered ratio']=(df['recovered_cases']/df['Total_cases'])*100
df.head()
```

```python
plt.figure(figsize=(10,4))
plt.plot(df['State'],df['Active_Ratio'],color='blue',marker='o',label='Active cases')
plt.plot(df['State'],df['Death_Ratio'],color='k',marker='*',label='Death cases')
plt.plot(df['State'],df['recovered ratio'],color='blue',marker='o',label='recovered')

plt.xticks(rotation=90,fontsize=13.5)
plt.yticks(fontsize=13.5)

plt.xlabel('State',fontsize=15)
plt.ylabel('Active Ratio',fontsize=15)

plt.title('Line plot ',fontsize=15)

plt.legend()
plt.show()
```
